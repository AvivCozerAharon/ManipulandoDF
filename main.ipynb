```python
# Importando as variaveis e lendo o banco direito do drive

import pandas as pd

from google.colab import  drive

drive.mount('/drive')

db = pd.read_csv("/drive/MyDrive/class_titanic.csv")
print(db['Age'])
```

```python
#Vendo os tipos das colunas no db

print(db.info())
```

```python
# Retirando os nulos de 'Age' e calculando a media.

db= db.dropna(subset=['Age'])

media = db['Age'].mean()
print(media.astype(int))
print(db['Age'])           
db.head()  
```

```python
# Filtrando diferenças da media (Parametro escolhido +- 12.5)

filtro  =db['Age'] > media-12.5
novodb = db[filtro]
filtro2 = novodb['Age'] < media+12.5
novodb = novodb[filtro2]
print(novodb['Age'])
```

```python
#Criando um novo dataframe com 5 colunas

dataframe = novodb[['Name','Age','PassengerId','Sex','Ticket']]

dataframe.head()
```

```python
#Criando uma nova coluna a partir de duas outras colunas
import random

dataframe.insert(1,'nova_coluna',(dataframe['Age']-dataframe['PassengerId'])*random.randint(0,1))
dataframe.head()
```

```python
# Embaralhando o dataframe

import numpy as np

dataframe_embaralhado = dataframe.reindex(np.random.permutation(dataframe.index))

dataframe_embaralhado.head()
```

```python
# Ordenando o dataframe pela coluna 'Age' (idade), utilizando o criterio de ascending(cresecente) em false, ou seja ordenado do maior ao menor


dataframe_ordenado = dataframe.sort_values(by='Age', ascending=False)
display(dataframe_ordenado)
```

```python
grupo2 = dataframe.groupby('Name')[['Age']].count()
print(grupo2)
```

```python
#Utilizando o map para converte a coluna sex de male/female para 0 e 1 boleanos, criando uma nova colna Sex_bollean

dataframe['Sex_bollean'] = dataframe['Sex'].map( {'female': 0,'male': 1})
dataframe.head()
```

```python
#Convertendo o dataframe para um csv e salvando no drive

dataframe.to_csv("/drive/MyDrive/titanic_novodf.csv")
```

```python
#Convertendo o dataframe para um json e salvando no drive

dataframe.to_json("/drive/MyDrive/titanic_novodf.json")
```
